### Data Analysis

```python
x_axis = ['age', 'bmi', 'children', 'expenses']

for x in x_axis:
    fig, axes = plt.subplots(1, 2, figsize=(18, 4))
    sns.distplot(df[x], ax = axes[0], kde = False)
    sns.boxplot(df[x], ax = axes[1], orient = "h", showmeans = True, color = "pink")
```

### Handling Categorical Datas

```python
df.head()
```

```python
df.sex.unique()
```

```python
df['sex'] = df.sex.map({'female': 0, 'male': 1})
```

```python
df.head()
```

```python
df['smoker'] = df.smoker.map({'yes': 1, 'no': 0})
```

```python
df.head()
```

```python
df.region.unique()
```

```python
df['region'] = df.region.map({'southwest': 1, 'southeast': 2, 'northwest': 3, 'northeast': 4})
```

```python
df.head()
```

```python
df.info()
```

### Split the dataset in x and y

```python
df.columns
```

```python
x = df.drop(['expenses'], axis = 1)
```

```python
x.head()
```

```python
y = df[['expenses']]
```

```python
y.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Model Training

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred1 = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score1 = r2_score(y_test, y_pred1)
```

```python
round(score1 * 100, 3)
```

### Hist Gradient Boosting Regressor

```python
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
hg = HistGradientBoostingRegressor()
```

```python
hg.fit(x_train, y_train)
```

```python
y_pred2 = hg.predict(x_test)
```

```python
score2 = r2_score(y_test, y_pred2)
```

```python
round(score2 * 100, 3)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred3 = rf.predict(x_test)
```

```python
score3 = r2_score(y_test, y_pred3)
```

```python
round(score3 * 100, 3)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred4 = dt.predict(x_test)
```

```python
score4 = r2_score(y_test, y_pred4)
```

```python
round(score4 * 100, 3)
```

### Gradient Boosting Regressor -> Best Model

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb = GradientBoostingRegressor()
```

```python
gb.fit(x_train, y_train)
```

```python
y_pred5 = gb.predict(x_test)
```

```python
score5 = r2_score(y_test, y_pred5)
```

```python
round(score5 * 100, 3)
```

### Voting Regressor

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
r1 = LinearRegression()
r2 = RandomForestRegressor(n_estimators = 10, random_state = 1)
r3 = KNeighborsRegressor()
```

```python
vc = VotingRegressor([('lr', r1), ('rf', r2), ('r3', r3)])
```

```python
vc.fit(x_train, y_train)
```

```python
y_pred6 = vc.predict(x_test)
```

```python
score6 = r2_score(y_test, y_pred6)
```

```python
round(score6 * 100, 3)
```
